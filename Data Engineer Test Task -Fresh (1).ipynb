Introduction
One of the key challenges that marketing teams must solve is allocating their resources in a way that 
minimizes “cost per acquisition” and increases return on investment. One of the major drivers for this is 
possible through segmentation, the process of dividing customers into different groups based on their 
behavior or characteristics.
Customer segmentation is to group customers who are similar to each other 

Resources: two file csv 
- Users
- Users Transactions

```python
#import libarys
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import FileLink
```

```python
#read csv file by pandas on local machine 
df_user = pd.read_csv("Users_202406231529.csv")
df_user.head(5)
```

```python
#read csv file by pandas on local machine 
df_user_Transactions= pd.read_csv("UsersTransactions_202406231512.csv")
df_user_Transactions.head(5)
```

```python
# to get unique values for column 'operation' 
df_user_Transactions.operator.unique()
```

### big issue that same value write with different format 
##### du = دو
##### etisalat = إتصالات = ETISALAT  =  etisalat |  

```python
# Define a mapping dictionary for standardizing operator names
operator_mapping = {
    'du': 'du',
    'دو': 'du',
    'etisalat': 'etisalat',
    'إتصالات': 'etisalat',
    'ETISALAT': 'etisalat',
    'etisalat |  ': 'etisalat'
}

# Replace values in 'operator' column using the mapping dictionary
df_user_Transactions['operator'] = df_user_Transactions['operator'].replace(operator_mapping)
```

```python
# check that problem is solved done 
df_user_Transactions.operator.unique()
```

```python
# Merging on the 'id', 'userid  column
merged_df = pd.merge(df_user, df_user_Transactions, left_on='id', right_on='userid', how='inner')
```

```python
# show dataset after join
merged_df.head(4)
```

```python
# check the column "amount" not have nulls or missing values 
merged_df['amount'].isnull().sum()
```

```python
#check data types of data
merged_df.dtypes
```

## I want to change data type column"date" object to datetime  2024-06-23 12:26:17  

```python
# Convert 'date' column to datetime
merged_df['date'] = pd.to_datetime(merged_df['date'])

# Extract date and time parts
merged_df['date_part'] = merged_df['date'].dt.date
merged_df['time_part'] = merged_df['date'].dt.time
```

```python
merged_df.head(5)
```

# 1) 3 segments :Divide people into 3 groups depending on "amount"

```python
# Calculate the minimum and maximum values of the 'amount' column
min_amount = merged_df['amount'].min()
max_amount = merged_df['amount'].max()

# Define the number of bins (segments) you want
num_segments = 3

# Calculate the bin width based on the number of segments
bin_width = (max_amount - min_amount) / num_segments

# Create the segments based on 'amount' column
segments = pd.cut(merged_df['amount'], bins=num_segments, labels=[f"Segment {i+1}" for i in range(num_segments)])

# Assign the segments back to the DataFrame
merged_df['segment'] = segments
```

```python
merged_df['segment'].unique()
```

```python
merged_df.head(5)
```

# 2) The average purchase transaction value

```python
# Calculate the average purchase transaction value in the "amount" column
average_amount = merged_df['amount'].mean()

print(f"The average purchase transaction value is: {average_amount:.2f}")
```

```python
merged_df.head(4)
```

# 3) - The top day for purchase

```python
# Convert 'date' column to datetime if it's not already
merged_df['date'] = pd.to_datetime(merged_df['date'])

# Extract the date part (excluding time)
merged_df['date_only'] = merged_df['date'].dt.date

# Count the number of purchases for each date
purchase_counts = merged_df['date_only'].value_counts()

# Identify the top day for purchases
top_day = purchase_counts.idxmax()
top_day_count = purchase_counts.max()

print(f"The top day for purchases is: {top_day} with {top_day_count} purchases")
```

# 4) The peak hour for purchase

```python
# Convert 'date' column to datetime if it's not already
merged_df['date'] = pd.to_datetime(merged_df['date'])

# Extract the hour part
merged_df['hour'] = merged_df['date'].dt.hour

# Count the number of purchases for each hour
purchase_counts_by_hour = merged_df['hour'].value_counts().sort_index()

# Identify the peak hour for purchases
peak_hour = purchase_counts_by_hour.idxmax()
peak_hour_count = purchase_counts_by_hour.max()

print(f"The peak hour for purchases is: {peak_hour}:00 with {peak_hour_count} purchases")
```

# 5)The top 10 Donors with values

```python
# Group by 'userid' and sum the 'amount' for each donor
donor_totals = merged_df.groupby('userid')['amount'].sum()

# Sort the donors by total transaction amount in descending order
top_donors = donor_totals.sort_values(ascending=False).head(10)

print("Top 10 Donors with their total transaction values:")
print(top_donors)
```
